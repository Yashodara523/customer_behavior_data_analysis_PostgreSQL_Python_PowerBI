## **Preprocess the Dataset**

```python
import pandas as pd
```

```python
df=pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
#if you want to categorycal columns also
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ' ,'_')
df.columns
```

```python
labels=['young_adult','adult','middle_age','senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
df.head()
```

```python
df['frequency_of_purchases'].unique()
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Annually': 365,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Monthly': 30,
    'Every 3 Months': 90
}

df['frequency_of_purchases'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df.head()
```

```python
df[['promo_code_used','discount_applied']].head()
```

```python
(df['discount_applied']==df['promo_code_used']).all()
```

```python
df = df.drop(columns=['promo_code_used'])
```
